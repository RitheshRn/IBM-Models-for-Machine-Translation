```python
from tqdm import tqdm
```

```python
# Read english sentences
with open('corpus.en', 'r') as f:
    english_corpus = f.readlines()
    english_corpus = [i[:-1] for i in english_corpus]
    english_corpus = [i.split(' ') for i in english_corpus]
    
# Read Spanish sentences
with open('corpus.es', 'r') as f:
    spanish_corpus = f.readlines()
    spanish_corpus = [i[:-1] for i in spanish_corpus]
    spanish_corpus = [i.split(' ') for i in spanish_corpus]
```

```python
# Initialize parameters

# english_vacab
english_vacab = {}
for sent in tqdm(english_corpus):
    for word in sent:
        english_vacab[word] = 0
english_vacab['NULL'] = 0

        
# get the counts n(e)
spanish_vocab = []
for word in tqdm(english_vacab):
    c = []
    for idx in range(len(english_corpus)):
        if word in english_corpus[idx]:
            c += spanish_corpus[idx]
    spanish_vocab += c
    english_vacab[word] = len(set(c))
english_vacab['NULL'] = len(set(spanish_vocab))


# compute t
t = {}
for s_idx in tqdm(range(len(english_corpus))):
    for e_word in english_corpus[s_idx]:
        for f_word in spanish_corpus[s_idx]:
            t[(s_idx, f_word, e_word)] = 1/english_vacab[e_word]
# account for NULL
for s_idx in tqdm(range(len(english_corpus))):
    for f_word in spanish_corpus[s_idx]:
        t[(s_idx, f_word, 'NULL')] = 1/english_vacab['NULL']
        
            
# compute cef 
cef = {}
for s_idx in tqdm(range(len(english_corpus))):
    for e_word in english_corpus[s_idx]:
        for f_word in spanish_corpus[s_idx]:
            cef[(s_idx, f_word, e_word)] = 0
for s_idx in tqdm(range(len(english_corpus))):
    for f_word in spanish_corpus[s_idx]:
            cef[(s_idx, f_word, 'NULL')] = 0
            
            
# compute ce
ce = {}
for s_idx in tqdm(range(len(english_corpus))):
    for e_word in english_corpus[s_idx]:
        ce[(s_idx, e_word)] = 0
    ce[(s_idx, 'NULL')] = 0
```

```python
for i in range(len(english_corpus)):
    english_corpus[i] += ['NULL']
```

```python
# IBM model 1

for iteration in tqdm(range(5)):
    # set counts to zero
    cef = dict.fromkeys(cef, 0)
    ce = dict.fromkeys(ce, 0)
    
    # iterate over corpus
    for k in range(len(english_corpus)):
        for i in range(len(spanish_corpus[k])):
            
            denom = sum([t[(k, spanish_corpus[k][i], english_corpus[k][z])] for z in range(len(english_corpus[k])-1)])
            
            for j in range(len(english_corpus[k])):
                
                # Compute delta
                num = t[(k, spanish_corpus[k][i], english_corpus[k][j])]
                delta = num/denom
                
                # Update
                cef[(k, spanish_corpus[k][i], english_corpus[k][j])] += delta
                ce[(k, english_corpus[k][j])] += delta
                
    # evelauate t
    for s_idx in tqdm(range(len(english_corpus))):
        for e_word in english_corpus[s_idx]:
            for f_word in spanish_corpus[s_idx]:
                t[(s_idx, f_word, e_word)] = cef[(s_idx, f_word, e_word)] /  ce[(s_idx, e_word)]
    
```

```python
# manipulate 'NULL' in t
for s_idx in tqdm(range(len(english_corpus))):
    for f_word in spanish_corpus[s_idx]:
        t[(s_idx, f_word, 'NULL')] = 0
```
